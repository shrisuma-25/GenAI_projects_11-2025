#### Python Data Structures

```python
collection = [
    {"title": "Inception", "genre": "Sci-Fi", "year": 2010},
    {"title": "The Matrix", "genre": "Action", "year": 1999},
    {"title": "Coco", "genre": "Animation", "year": 2017},
]
# print all movies in the collection
for movie in collection:
    # Use an f-string to format the output for each movie dictionary
    print(f'"{movie["title"]}" is a  {movie["genre"]} movie released in ({movie["year"]})')

# Ask user for a genre to search
search_genre = input("Enter a genre to search for: ")
# Find and print movies that match the search genre
print(f'\nMovies in the genre "{search_genre}":')
found = False
for movie in collection:
    if movie["genre"].lower() == search_genre.lower():
        print(f'- {movie["title"]} ({movie["year"]})')
        found = True
if not found:
    print("No movies found in that genre.")

  

```

#### Basic List Operations

```python
# Create list of 5 fruits
fruits = ["apple", "strawberry", "pineapple", "mango", "banana"]
print("First fruit:", fruits[0])          
print("Last fruit:", fruits[-1])    
#add new fruit to the list
fruits.append("orange")
print(fruits)

#  Remove a 3rd fruit from the list  
fruits.pop(2)
print(fruits)   

#Sort the list alphabetically
fruits.sort()
print("Sorted fruits:", fruits)
#sorted(fruits)
```

Loops with Lists

```python
numbers = [3, 6, 2, 8, 4]

for num in numbers:
    print(num)

#Calculate and print the sum of all numbers in the list
total_sum = 0
for num in numbers:
    total_sum += num
print("Sum of numbers:", total_sum)

# List of multiplie numbers by 2
multiplied_numbers = []
for num in numbers:
    multiplied_numbers.append(num * 2)
print("Multipled list numbers:", multiplied_numbers)
```

#### List Comprehensions

```python
squares_numbers = [x**2 for x in range(1, 11)]
print("Squares of numbers from 1 to 10:", squares_numbers)

# List of even numbers from 1 to 10
even_numbers = [x for x in range(1, 11) if x % 2 == 0]
print("Even numbers from 1 to 10:", even_numbers)
```

#### Nested Lists

```python
matrix = [[1,2], [3,4], [5,6]]
#print second sublist
print("Second sublist:", matrix[1])
#access the number 4 from the nested list
print("print number 4 from matrix :", matrix[1][1])
# Print all elements in the nested list
for sublist in matrix:  
    for element in sublist:
        print(element)
```

```python
scores =[78, 85, 90, 56, 99, 74]
# Calculate and print the average score
total_score = sum(scores)
average_score = total_score / len(scores)
print("Average score:", average_score)
# Print scores >80
print("Scores greater than 80:")
for score in scores:
    if score > 80:
        print(score)
#replace score less than 60 with Fail
for i in range(len(scores)):
    if scores[i] < 60:
        scores[i] = "Fail"
print("Updated scores:", scores)
```
