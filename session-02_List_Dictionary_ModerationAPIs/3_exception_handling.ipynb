#### Exception Handling


```python
person = {
    "name": "Alice",
    "age": 30,
    "is_student": False
}
```

```python
# Try Catch
# Try Except
try:
    del person[is_student] 
except Exception as e:
    print(f"Error in removing key: {e}")
print(person)
```

```python
try:
    a = 10
    b = 0
    print(a / b)
except Exception as e:
    print(f"Error occurred: {e}") 
    
```
