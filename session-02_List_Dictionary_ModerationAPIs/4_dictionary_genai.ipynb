```python
#key value pair
message_history = {"user_prompt": "When does it rain in California?", "system_response": "California typically experiences its rainy season from November to March."}
print(message_history["system_response"])
message_history.get("system_response")  # Accessing value using get() method
```

```python
message_user = {"role": "user", "content": "When does it rain in California?"}
message_system = {"role": "system", "content": "California typically experiences its rainy season from November to March."}
#list of dictionary
messages = [message_user, message_system] # Key Value Pair

i = 0
for message in messages:
    print(f" {i+1}: {message}") 
    i += 1
```
