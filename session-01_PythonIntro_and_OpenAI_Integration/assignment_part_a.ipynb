#### Reverse a String

```python
def reverse_string(s):
    reversed_s = ""
    for char in s:
        reversed_s = char + reversed_s
    return reversed_s

my_string = "welcome"
my_reverse_string = reverse_string(my_string)
print(my_reverse_string)
```

#### Check for palindrome

```python
def is_palindrome(s):
    # Normalize: remove spaces and convert to lowercase
    cleaned = ""
    for char in s:
        if char != " ":
            cleaned += char.lower()
    
    # Check palindrome manually without using [::-1]
    left = 0, 
    right = len(cleaned) - 1
    
    while left < right:
        if cleaned[left] != cleaned[right]:
            return False
        left += 1
        right -= 1
    
    return True
print(is_palindrome("Race car"))       # True
print(is_palindrome("Malyalam"))  # False

```

#### Find the Missing Number

```python
list_of_numbers = [1, 2, 4, 5, 6]
def find_missing_number(nums):
    n = len(nums) + 1  # Since one number is missing
    expected_sum = n * (n + 1) // 2
    actual_sum = 0
    for num in nums:
        actual_sum += num
    return expected_sum - actual_sum
missing_number = find_missing_number(list_of_numbers)
print(missing_number)  # Output: 3
```

#### Count Word frequency

```python
def word_frequency(text):
    frequency = {}
    words = text.split() # Split text into words
    for word in words:
        word = word.lower()  # Normalize to lowercase
        if word in frequency:
            frequency[word] += 1
        else:
            frequency[word] = 1
    return frequency
print(word_frequency("This is a test This test is only a test"))
print(word_frequency("AI makes AI easier"))
```

#### FizzBuzz

```python
def fizzbuzz():
    for i in range(1, 31):
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
        elif i % 3 == 0:
            print("Fizz")
        elif i % 5 == 0:
            print("Buzz")
        else:
            print(i)
```
