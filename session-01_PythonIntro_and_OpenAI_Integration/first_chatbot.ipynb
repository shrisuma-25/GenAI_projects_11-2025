```python
import os
from dotenv import load_dotenv
load_dotenv(override=True, dotenv_path="../.env")
my_api_key = os.getenv("OPENAI_API_KEY")

print(my_api_key[:5] + "..." + my_api_key[-5:])  # Print first and last 5 characters of the API key for verification
```

```python
#pip install openai
from openai import OpenAI
client = OpenAI(api_key=my_api_key)

def ask_question(prompt):
    context = ""
    print(f"User asked: {prompt}")
    response = client.chat.completions.create(
        model="gpt-5-nano",
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": context + prompt}
        ]
    )
    print (response)
    return response.choices[0].message.content
```

```python
#user_prompt = "Ask Something."
user_prompt = "Who was Napolean Bonaparte."

response = ask_question(user_prompt)
print("\nOpenAI says:", response)
```

```python
import time
while True:
    # Ask user for a question
    user_prompt = input("Ask something: ")

    if (user_prompt.lower() != 'quit'):
        # Get and print the response
        response = ask_question(user_prompt)
        print("\nOpenAI says:", response)
        
        # add delay of 3 seconds
        time.sleep(3)
    else:
        print("Exiting the loop. Goodbye!")
        break    
```
